## Merge

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1 # 고객 목록
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2 # 구매 기록
```

```python
# pandas.merge(left, right)
# pd.merge(df1, df2)
pd.merge(left=df1, right=df2, how='inner', on='고객번호')  # 이름, 금액 모두 고객번호에 해당하는게 존재하는 것만 합쳐주겠다
```

```python
pd.merge(left=df1, right=df2, how='outer', on='고객번호')
```

```python
pd.merge(left=df1, right=df2, how='left', on='고객번호')  # left 존재하는 모든 행의 데이터 (기준 열), right에 nan이 존재해도 그 상태로 받아오겠다
```

```python
pd.merge(left=df1, right=df2, how='right', on='고객번호')
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]

df1 = pd.DataFrame(data)
df1
```
